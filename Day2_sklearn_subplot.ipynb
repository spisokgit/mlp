```python
# sklearn AT&T 사람 얼굴 데이터
from sklearn import datasets
olivetti = datasets.fetch_olivetti_faces()
X = olivetti.images
y = olivetti.target
print(X.shape)

# 사람 얼굴 데이터 시각화
import matplotlib.pyplot as plt
fig, axes = plt.subplots(4, 10, figsize=(20,8))
for i in range(4):
    for j in range(10):
        axes[i, j].imshow(X[i*10+j], cmap='gray')
        axes[i, j].set_title(str(y[i*10+j]))
        axes[i, j].axis('off')
plt.subplots_adjust(wspace=0, hspace=0.2)
plt.show()
```
